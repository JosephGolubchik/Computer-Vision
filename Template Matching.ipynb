```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 6]
```

```python
img = cv2.imread('pyramids.png',0)
plt.imshow(img,'gray')
```

```python
template = img[420:520,210:310]
template = template - np.mean(template)
plt.imshow(template,'gray')
```

```python
res = cv2.filter2D(img-img.mean(),cv2.CV_64F,template)
# res2 = np.sqrt(cv2.filter2D(img**2,cv2.CV_64F,template**2))
# res3 = res/res2
plt.imshow(res,'gray')
```
